```python
import pandas as pd
import statsmodels.api as sm
import matplotlib.pyplot as plt
df = pd.read_csv('subfiles/sub-data.csv')
df.head()
```

```python
plt.figure(figsize = (20,20))
plt.ylabel('Annual Household Income')
plt.xlabel('Number of Kids')
plt.title('THE GRAPH', font='Times New Roman')
plt.scatter(df['KIDS'], df['INCOME'])
```

```python
x_ary = df['KIDS']
xConstant = sm.add_constant(x_ary)
model = sm.OLS(df['INCOME'], xConstant)
results = model.fit()
b,m = results.params
print(m)
print(b)
plt.figure(figsize = (20,20))
plt.scatter(df['KIDS'], df['INCOME'], color='Purple', alpha=0.7)
plt.yscale('log')
plt.plot(x_ary, m*x_ary+b, color='Navy')
```

## The logarithmic graph demonstrates that most of the points are "squished" down near the x-axis. While this offers an approximate mean of the data, it is clear that the relationship between number of kids and income is not correlated and therefore not causal (duh)
